```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_data = pd.read_csv (r'C:\Users\HP\Documents\notebook\DOWNLOADS\spaceship-titanic\train.csv')
train_data.head()
```

```python
test_data = pd.read_csv(r'C:\Users\HP\Documents\notebook\DOWNLOADS\spaceship-titanic\test.csv')
test_data.head()
```

```python
y = train_data['Transported']

Attributes = ['CryoSleep','VIP']

x = pd.get_dummies(train_data[Attributes])
x_test = pd.get_dummies(test_data[Attributes])

model = RandomForestClassifier(n_estimators = 100, max_depth = 6, random_state =1)

model.fit(x,y)
predictions = model.predict(x_test)
output = pd.DataFrame({'PassengerId':test_data.PassengerId, 'Transported':predictions})
output.to_csv ('SPACESHIP.csv', index = False)
print ('successfully done')
```
